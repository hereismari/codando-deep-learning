## Motivação

Vários expecialistas na área defendem a ideia de que a melhor forma de aprender Deep Learning é **na prática**. Isto é, na medida que vamos sendo expostos a novos conceitos implementá-los nos mínimos detalhes para de fato entende-los.

> *By far the best strategy for learning a lot that I discovered was to reimplement everything. I happened to do a lot of it in Javascript for lols. Whenever I read something I fall into an illusion of understanding it and forcing myself to reimplement it always uncovers interesting insights. It’s my favorite way to learn.* - [Andrej Kaparthy](https://www.quora.com/What-was-your-path-toward-ML-DL-What-books-did-you-enjoy-most-while-learning-DL)

> *This is the most extensive list of &quot;From Scratch Machine Learning&quot; I&#39;ve ever seen - this is a golden resource if your learn like I do (by building things from scratch). [https://t.co/LcoPAQBULv](https://t.co/LcoPAQBULv)* - [Trask (@iamtrask)](https://twitter.com/iamtrask/status/1034376270128267264?ref_src=twsrc%5Etfw)

A verdade é que ter acesso ao ensino de Deep Learning se tornou extremamente fácil, porém agora o problema é outro: "por onde devo seguir? livros? vídeos? artigos? frameworks?". Aqui iremos tomar a abordagem de ensinar conceitos de Deep Learning **para programadores** da maneira que eles/elas já estão habituados: **programando do zero!**. Aprender um conceito de maneira profunda (tun dun tisss) ainda é difícil e uma tarefa que exige rigor e disciplina.

Esse site busca facilitar sua tragetória, indicando sugestões de materiais, atividades, papers, vídeos para cada assunto apresentado junto de **código python implementando cada conceito em detalhes**.
